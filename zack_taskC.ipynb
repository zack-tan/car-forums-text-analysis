```python
#defining the text processing method to eliminate punctuation and stopwords
import string

#importing stopwords from the natural language toolkit library 
from nltk.corpus import stopwords

def text_process(mess):
    nopunc =[char for char in mess if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]

df['Message_words'] = df['Message'].apply(text_process)
df
```

```python
# Read in CSV file for models
df_models = pd.read_csv("models.csv",header=None)


```

### Remove all occurrences of brands and labels from the corpus

```python
# Add all the brands and makes into a list
wordlist_raw = [brand for brand in df_models[0]]
wordlist_raw.append([model for model in df_models[1]])
wordlist = []
for t in wordlist_raw:
    if t not in wordlist:
        wordlist.append(t)
wordlist
```

```python
# Remove brands and makes
def remove_brands(text):
    filtered = []
    for t in text:
        if (not any(t.lower() in word for word in wordlist)) and (not any(t in word for word in wordlist)):
            filtered.append(t)    
    return filtered

df['Message_words'] = df['Message_words'].apply(remove_brands)
```
