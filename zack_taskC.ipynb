```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

df = pd.read_pickle("data_clean.pkl")

df
```

```python
# Read in CSV file for models
df_models = pd.read_csv("models.csv",header=None)
```

### Remove all occurrences of brands and labels from the corpus

```python
# Add all the brands and makes into a list
wordlist_raw = [brand for brand in df_models[0]]
wordlist_raw.append([model for model in df_models[1]])
wordlist = []
for t in wordlist_raw:
    if t not in wordlist:
        wordlist.append(t)
wordlist
```

```python
# Remove brands and makes
def remove_brands(text):
    filtered = []
    for t in text:
        if (not any(t.lower() in word for word in wordlist)) and (not any(t in word for word in wordlist)):
            filtered.append(t)    
    return filtered

df['Message_words'] = df['Message_words'].apply(remove_brands)
```

```python
df
```

```python
# Collate all text into list
alltext = []
for comment in df['Message_words']:
    for text in comment:
        alltext.append(text)
alltext
```

```python
# Count frequencies
freqDist = nltk.FreqDist(text.lower() for text in alltext)
```

```python
# Convert to df for visualization
df_freq = pd.DataFrame.from_dict(freqDist, orient='index', columns=['freq'])
df.reset_index(level=0, inplace=True)
df_freq.sort_values(by=['freq'], inplace=True, ascending=False)

pd.set_option('display.max_rows', 100)
df_freq.head(100)
```
